# Matriks Transformasi

Matriks transformasi adalah alat matematika yang digunakan untuk memetakan satu titik atau vektor ke posisi baru dalam ruang koordinat. Dalam konteks aljabar linear dan grafika komputer, matriks ini memungkinkan kita untuk melakukan manipulasi objek secara efisien.Secara umum, sebuah transformasi linear terhadap vektor $\mathbf{v}$ dapat dinyatakan dalam persamaan:$$\mathbf{v'} = M \cdot \mathbf{v}$$Di mana:$\mathbf{v}$ adalah vektor input (posisi awal).$M$ adalah matriks transformasi.$\mathbf{v'}$ adalah vektor output (posisi akhir).

## contoh soal
Titik $A(1, 2)$
Transformasi dari:
- A ke B
- B ke C
- C ke D
- D ke E
- E ke F

1. Transformasi $A \to B$ (Skala)
Titik $A$ diperbesar dengan faktor skala $s_x = 2$ dan $s_y = 2$.
$$\begin{aligned}
\mathbf{B} &= \mathbf{M}_{scale} \cdot \mathbf{A} \\
\begin{bmatrix} x_B \\ y_B \\ 1 \end{bmatrix} &= \begin{bmatrix} 2 & 0 & 0 \\ 0 & 2 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 1 \\ 2 \\ 1 \end{bmatrix} = \begin{bmatrix} 2 \\ 4 \\ 1 \end{bmatrix}
\end{aligned}$$
Hasil: Titik $B$ berada di $(2, 4)$.

2. Transformasi $B \to C$ (Rotasi $90^\circ$)Titik $B$ diputar sejauh $90^\circ$ berlawanan arah jarum jam terhadap titik asal.$$\begin{aligned}
\mathbf{C} &= \mathbf{M}_{rot} \cdot \mathbf{B} \\
\begin{bmatrix} x_C \\ y_C \\ 1 \end{bmatrix} &= \begin{bmatrix} 0 & -1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 2 \\ 4 \\ 1 \end{bmatrix} = \begin{bmatrix} -4 \\ 2 \\ 1 \end{bmatrix}
\end{aligned}$$Hasil: Titik $C$ berada di $(-4, 2)$.

3. Transformasi $C \to D$ (Translasi)Titik $C$ digeser sejauh $t_x = 5$ dan $t_y = 3$.$$\begin{aligned}
\mathbf{D} &= \mathbf{M}_{trans} \cdot \mathbf{C} \\
\begin{bmatrix} x_D \\ y_D \\ 1 \end{bmatrix} &= \begin{bmatrix} 1 & 0 & 5 \\ 0 & 1 & 3 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} -4 \\ 2 \\ 1 \end{bmatrix} = \begin{bmatrix} 1 \\ 5 \\ 1 \end{bmatrix}
\end{aligned}$$Hasil: Titik $D$ berada di $(1, 5)$.

4. Transformasi $D \to E$ (Refleksi Sumbu-Y)Titik $D$ dicerminkan terhadap sumbu vertikal (Y).$$\begin{aligned}
\mathbf{E} &= \mathbf{M}_{refl} \cdot \mathbf{D} \\
\begin{bmatrix} x_E \\ y_E \\ 1 \end{bmatrix} &= \begin{bmatrix} -1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 1 \\ 5 \\ 1 \end{bmatrix} = \begin{bmatrix} -1 \\ 5 \\ 1 \end{bmatrix}
\end{aligned}$$Hasil: Titik $E$ berada di $(-1, 5)$.

5. Transformasi $E \to F$ (Shear Horizontal)Titik $E$ diberikan efek geser (shear) pada sumbu X dengan faktor $k = 2$.$$\begin{aligned}
\mathbf{F} &= \mathbf{M}_{shear} \cdot \mathbf{E} \\
\begin{bmatrix} x_F \\ y_F \\ 1 \end{bmatrix} &= \begin{bmatrix} 1 & 2 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} -1 \\ 5 \\ 1 \end{bmatrix} = \begin{bmatrix} 9 \\ 5 \\ 1 \end{bmatrix}
\end{aligned}$$Hasil: Titik $F$ berada di $(9, 5)$.

### Tabel Ringkasan Rantai Transformasi

| Tahap | Operasi | Matriks Transformasi | Titik Input | Titik Output |
| :--- | :--- | :--- | :--- | :--- |
| **$A \to B$** | Skala (2, 2) | $\begin{pmatrix} 2 & 0 & 0 \\ 0 & 2 & 0 \\ 0 & 0 & 1 \end{pmatrix}$ | $(1, 2)$ | **$(2, 4)$** |
| **$B \to C$** | Rotasi $90^\circ$ | $\begin{pmatrix} 0 & -1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & 1 \end{pmatrix}$ | $(2, 4)$ | **$(-4, 2)$** |
| **$C \to D$** | Translasi (5, 3) | $\begin{pmatrix} 1 & 0 & 5 \\ 0 & 1 & 3 \\ 0 & 0 & 1 \end{pmatrix}$ | $(-4, 2)$ | **$(1, 5)$** |
| **$D \to E$** | Refleksi Sumbu-Y | $\begin{pmatrix} -1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}$ | $(1, 5)$ | **$(-1, 5)$** |
| **$E \to F$** | Shear-X ($k=2$) | $\begin{pmatrix} 1 & 2 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}$ | $(-1, 5)$ | **$(9, 5)$** |

```python
# Create a final GeoGebra .ggb file that matches the full transformation table A→F

import zipfile

ggb_path = "transformasi_final.ggb"

xml_content = """<?xml version="1.0" encoding="utf-8"?>
<geogebra format="5.0" version="5.0.0.0">
  <construction title="Transformasi Lengkap A-F" author="ChatGPT">

    <!-- Titik awal -->
    <element type="point" label="A">
      <coords x="1" y="2" z="1"/>
    </element>

    <!-- Scaling -->
    <command name="Dilate">
      <input a0="A" a1="2"/>
      <output a0="B"/>
    </command>

    <!-- Rotasi 90 derajat (pi/2 rad) -->
    <element type="numeric" label="angle">
      <value val="1.57079632679"/>
    </element>

    <command name="Rotate">
      <input a0="B" a1="angle"/>
      <output a0="C"/>
    </command>

    <!-- Translasi (5,3) -->
    <element type="vector" label="v">
      <coords x="5" y="3" z="0"/>
    </element>

    <command name="Translate">
      <input a0="C" a1="v"/>
      <output a0="D"/>
    </command>

    <!-- Refleksi sumbu-Y -->
    <expression label="E" exp="(-x(D), y(D))"/>

    <!-- Shear-X k=2 -->
    <expression label="F" exp="(x(E)+2*y(E), y(E))"/>

  </construction>
</geogebra>
"""

with zipfile.ZipFile(ggb_path, 'w') as z:
    z.writestr("geogebra.xml", xml_content)

ggb_path
```
